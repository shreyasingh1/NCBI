```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
%matplotlib inline

#hiding warning messages
import warnings
warnings.filterwarnings("ignore")
import time 
```

# NB

```python
#Reading in grep files
grep_nb = pd.read_csv('grep_nb.txt', sep = '\t')
grep_nb = grep_nb.T
grep_nb = grep_nb.reset_index()
grep_nb = grep_nb.T
grep_nb = grep_nb.reset_index(drop = True)

#converting to str (NB)
grep_nb = grep_nb.astype(str)

#splitting scores 
str_split_nb = grep_nb[1].tolist()
i = 0
while (i < len(str_split_nb)):
    str_split_nb[i] = str_split_nb[i].split()
    i = i + 1
    
#making df
scores_nb = pd.DataFrame(str_split_nb)

#label df
labels_nb = pd.DataFrame(grep_nb[0])
labels_nb = labels_nb.rename(columns = {0 : 'labels'})

#combining
data_nb = pd.concat([labels_nb, scores_nb], axis = 1)

#reindexing
data_nb = data_nb.rename(columns = {0 : 'ACC', 1 : 'AUC', 2 : 'MCC', 3 : 'SENS', 4 : 'SPHE', 5 : 'BACC',})
data_nb = data_nb.set_index('labels')

#seperating datasets into Train, Test, Ext1, Ext2, Ext3
data_nb = data_nb.iloc[:25]
data_nb.loc['output_maccs:Test']
subset_nb_train = data_nb.loc[['output_maccs:Train', 'output_top:Train', 'output_ecfp:Train', 'output_fcfp:Train', 'output_pub:Train',], :]
subset_nb_test = data_nb.loc[['output_maccs:Test', 'output_top:Test', 'output_ecfp:Test', 'output_fcfp:Test', 'output_pub:Test',], :]
subset_nb_ext1 = data_nb.loc[['output_maccs:Ext1', 'output_top:Ext1', 'output_ecfp:Ext1', 'output_fcfp:Ext1', 'output_pub:Ext1',], :]
subset_nb_ext2 = data_nb.loc[['output_maccs:Ext2', 'output_top:Ext2', 'output_ecfp:Ext2', 'output_fcfp:Ext2', 'output_pub:Ext2',], :]
subset_nb_ext3 = data_nb.loc[['output_maccs:Ext3', 'output_top:Ext3', 'output_ecfp:Ext3', 'output_fcfp:Ext3', 'output_pub:Ext3',], :]
```
